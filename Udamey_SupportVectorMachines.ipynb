```python
# our libarys 
#Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Note, we won't use seaborn due to a plot at the end.
```

```python
from sklearn import datasets 
# Load Iris data set 
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
# Show the Built-in Data Description
print(iris.DESCR)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.model_selection import train_test_split
# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.35,random_state = 3)
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn import metrics
```

```python
predicted = model.predict(X_test)

expected= Y_test
```

```python
print(metrics.accuracy_score(expected,predicted))
```

```python
# import all os SCM 
from sklearn import svm

X = iris.data[:,:2]
Y = iris.target
```

```python
C = 1.0
svc = svm.SVC(kernel='linear',C=C).fit(X,Y)
```

```python
rbf_svc = svm.SVC(kernel='rbf',gamma=0.7,C=C).fit(X,Y)
```

```python
poly_svc = svm.SVC(kernel='poly',degree=3,C=C).fit(X,Y)
```

```python
lin_svc = svm.LinearSVC(C=C).fit(X,Y)
```

```python
h = 0.02 

x_min = X[:,0].min() - 1
x_max = X[:,0].max() + 1

y_min = X[:,1].min() - 1
y_max = X[:,1].max() + 1
```

```python
xx, yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```python
# title for the plots
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

```python
# Use enumerate for a count
for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
    
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    plt.figure(figsize=(15,15))
    # Set the subplot position (Size = 2 by 2, position deifined by i count
    plt.subplot(2, 2, i + 1)
    
    # SUbplot spacing
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    
    # Define Z as the prediction, not the use of ravel to format the arrays
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    
    # Contour plot (filled with contourf)
    plt.contourf(xx, yy, Z, cmap=plt.cm.terrain, alpha=0.5,linewidths=0)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Dark2)
    
    # Labels and Titles
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
    


#plt.show()
```
